```python
import pandas as pd
```

```python
df = pd.read_csv("job_postings.csv")
```

```python
df.head()
```

```python
new_df = df[["title","description"]]
new_df["data"] = df["title"].astype("str")+" "+df["description"]
new_df = pd.DataFrame(new_df["data"])
new_df = new_df.dropna()
```

```python
new_df.head()
```

```python
from nltk.corpus import stopwords
# import nltk
# nltk.download('stopwords')
stop_words = stopwords.words('english')
stop_words.append('nan')
stop_words.append('removed')
stop_words.remove('not')
stop_words.remove('no')
stop_words[:10]

new_df['data'] = new_df['data'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
new_df.head()
```

```python
new_df.to_csv("job_postings.csv")
```
