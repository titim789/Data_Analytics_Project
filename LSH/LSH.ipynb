```python
def generate_hash_function(k,vector_dimensions):
    hash_functions_list = []
    curr_k = 0
    while curr_k < k:

        hash_function = []
        for j in range(vector_dimensions):    
            hash_value = random.randint(-1, 1) # CHANGED FROM -3 to 3 
            hash_function.append(hash_value)
        if (hash_function in hash_functions_list):
            continue
        hash_functions_list.append(hash_function)
        curr_k+=1

    return hash_functions_list
```

```python
random_hash_functions = generate_hash_function(k,vector_dimensions)
random_hash_functions_df = pd.DataFrame({'Hash_Functions': random_hash_functions})
```

```python
def generate_binary_vectors(input_vector,random_hash_functions_df):
    binary_vector = []
    for i,row in random_hash_functions_df.iterrows():
        input_vector = np.asarray(input_vector, dtype=np.float64)
        hash_function = np.asarray(row['Hash_Functions'], dtype=np.float64)
    
        dot_product = np.dot(input_vector,hash_function)
        # result = 0 if dot_product <= 0 else 1
        result = np.where(dot_product <= 0, 0, 1)
        binary_vector.append(result.tolist())
    return binary_vector
```

```python
def cosine_sim(v1,v2):
    if (len(v1) != len(v2)):
        print("Vectors not of the same dimension")
        return -1
    dotProduct = 0
    sumOfSquaredV1 = 0
    sumOfSquaredV2 = 0
    for i, pt in enumerate(v1):
        dotProduct += pt*v2[i]
        sumOfSquaredV1 += pt*pt
        sumOfSquaredV2 += v2[i] * v2[i]
    if dotProduct != 0:
        cosineSimilarity = dotProduct/((sumOfSquaredV1 ** 0.5) * (sumOfSquaredV2 ** 0.5))
    else:
        cosineSimilarity = 0
    return cosineSimilarity
```

```python
#Create new dataset

n = 500
dims = 5
query = []

# Create query vector
for i in range(dims): 
    query.append(random.randint(-1,1))

# Create random hashes of (n,dims)
dataset = generate_hash_function(n,dims)
```

```python
def generate_binary_vectors_list_input(input_vector,random_hash_functions):
    binary_vector = []
    for hash_function in random_hash_functions:
        input_vector = np.asarray(input_vector, dtype=np.float64)
        hash_function = np.asarray(hash_function, dtype=np.float64)
    
        dot_product = np.dot(input_vector,hash_function)
        # result = 0 if dot_product <= 0 else 1
        result = np.where(dot_product <= 0, 0, 1)
        binary_vector.append(result.tolist())
    return binary_vector
```

```python
for k in range(10,1600,20):

    # Create random hash functions
    hash_functions = generate_hash_function(k,dims)

    # Dot Product
    query_binary_vectors = generate_binary_vectors_list_input(query,hash_functions)
    dataset_binary_vectors = []
    for data in dataset:
        dataset_binary_vectors.append(generate_binary_vectors_list_input(data,hash_functions))

    # print(query_binary_vectors)
    # print(len(dataset_binary_vectors))

    mismatched_list = []

    for data in dataset_binary_vectors: 

        mismatched_bits = 0

        for i in range(k):

            if query_binary_vectors[i] != data[i]:
                mismatched_bits += 1

        mismatched_list.append(mismatched_bits)
    
    # print(len(mismatched_list))

    df = pd.DataFrame(data = {'Data': dataset ,'Distance': mismatched_list})

    # cosine similiarity
    df['Linear Vectors'] = df['Data'].apply(lambda x: cosine_sim(query,x))

    #Sort 
    linear = df.sort_values('Linear Vectors', ascending=False, ignore_index=True)
    lsh = df.sort_values('Distance', ignore_index=True)

    count = 0
    for i, row in linear.iterrows():
        if row['Data'] != lsh.iloc[i]["Data"]:
            count += 1


    print(f"[{k},{count}]", end=", ")
```
